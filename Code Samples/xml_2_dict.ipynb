# we need to generate some data

```python
# fun helps
def get_cat(N=3):
    '''returns a random ALPHA code from N possible codes'''
    letters = string.ascii_uppercase
    L = len(letters)
    if N > L**2:
        print('This definition need more logic to support sample space, N, '
              'grater than %s^2.\nSetting N=%s' % (L, L**2))
        N = L**2
    n = random.randint(1,N)
    return (string.ascii_uppercase[n-1] if (n < L) else
            '%s%s' % (string.ascii_uppercase[floor(n/L)-1],string.ascii_uppercase[(n%L)-1]))


def get_gender():
    '''returns a random gender'''
    return ['Male', 'Female'][random.randint(0,1)]

    
def get_id():
    '''returns a random universally unique identifier'''
    return str(uuid.uuid4()).replace('-','')

    
def mk_row(N=3):
    '''makes a random row of example data'''
    return [get_cat(N=N), get_gender(), get_id()]
    

def mk_example_df(I=10,N=3):
    '''makes a random dataframe of data for this example.
    input: I - number of rows you want
           N - number of categories in Cat variable column
           '''
    return pd.DataFrame([mk_row(N=N) for i in range(I)],
                        columns = ['Cat','Gender','id'])

# show how it works
mk_example_df(I=10)
```

```python
# making example data
example_df = mk_example_df(I=100,N=10)
example_df
```

# Now we will wright our example data

```python
def dataframe_2_xml(df,file_path_str):
    '''wright df to file_path_srt as .xml'''
    if not '.xml' in file_path_str:
        raise Exception("Sorry, file_path_str must have .xml extension")
    elif type(df) != type(pd.DataFrame([])):
        raise TypeError("Sorry, df must be a pandas DataFrame")
    else:
        df.to_xml(file_path_str, index=False)
```

this is an example of reading in .xlm file and returning a dictionary

```python
def read_xml_2_dict(file_path_str):
    '''reading in .xlm file and returning a dictionary'''
    if not '.xml' in file_path_str:
        raise Exception("Sorry, file_path_str must have .xml extension")
    else:
        return pd.read_xml("example_df.xml").to_dict()

def dict_wright(dict, file_path_str):
    '''wrights dictionary as json data'''
    if not '.json' in file_path_str:
        raise Exception("Sorry, file_path_str must have .json extension")
    else:
        with open(file_path_str, 'w') as f:
            json.dump(dict, f)
```

```python
new_dict = read_xml_2_dict("example_df.xml")
dict_wright(new_dict, "new_dict.json")
```
